```python
import networkx as netx
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (13,9)

```

The first thing I need to do is create a graph object. See below:

```python
go = netx.Graph()
```

Afterwards I will add noeds to the graph. See below:

```python
go.add_nodes_from(["Andre","Beverly","Ed","Carol","Diane","Fernando","Garth","Heather","Ike","Jane"])
```

I will now create relationships between the nodes. See below:

```python


go.add_edges_from([("Andre","Beverly"),("Andre","Carol"),("Andre","Diane"),("Andre","Fernando"),
                  ("Beverly","Diane"),("Beverly","Ed"),("Beverly","Garth"),
                  ("Carol","Diane"),("Carol","Fernando"),
                  ("Diane","Ed"),("Diane","Fernando"),("Diane","Garth"),
                  ("Ed","Garth"),
                  ("Fernando","Garth"),("Fernando","Heather"),
                  ("Garth","Heather"),
                  ("Heather","Ike"),
                  ("Ike","Jane")])

```

In my final step, I will draw the graph. See below:

```python


netx.draw_networkx(go)

```
